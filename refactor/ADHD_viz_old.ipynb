```python
summary_data_ctrl_adhd4 = ctrl_subjects.groupby(['community', 'ADHD4']).size().reset_index(name='count').query("ADHD4 == 'Ctrl'").drop(columns='ADHD4')
summary_data_ctrl_adhd4['Ctrl'] = 'Ctrl'

summary_data_ADHD1 = data.groupby(['community', 'ADHD1']).size().reset_index(name='count').query("ADHD1 == 'ADHD'").assign(ADHD1='ADHD1')
summary_data_ADHD2 = data.groupby(['community', 'ADHD2']).size().reset_index(name='count').query("ADHD2 == 'ADHD'").assign(ADHD2='ADHD2')
summary_data_ADHD3 = data.groupby(['community', 'ADHD3']).size().reset_index(name='count').query("ADHD3 == 'ADHD'").assign(ADHD3='ADHD3')
summary_data_ADHD4 = data.groupby(['community', 'ADHD4']).size().reset_index(name='count').query("ADHD4 == 'ADHD'").assign(ADHD4='ADHD4')
```

```python
for col in ['ADHD1', 'ADHD2', 'ADHD3', 'ADHD4', 'Ctrl']:
    if col not in summary_data_combined.columns:
        summary_data_combined[col] = pd.NA

summary_data_combined
```

```python
agg_data_test = summary_data_combined.groupby(['community', 'ADHD1', 'ADHD2', 'ADHD3', 'ADHD4','Ctrl'], dropna=False)['count'].sum().reset_index()
agg_data_test
```

```python
agg_data_tidy = agg_data_test.melt(id_vars=['community', 'count'], value_vars=['ADHD1', 'ADHD2', 'ADHD3', 'ADHD4', 'Ctrl'], 
                                   var_name='ADHD_label', value_name='ADHD').dropna().drop(columns='ADHD')
agg_data_tidy
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='community', y='count', hue='ADHD_label', data=agg_data_tidy, palette='Set3')
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
plt.xlabel('Community')
plt.ylabel('Count')
plt.title('ADHD Label Distribution by Community')
plt.show()
```

```python
proportion_data = agg_data_tidy.groupby('community').apply(
    lambda df: df.assign(group_total=df['count'].sum(), proportion=df['count'] / df['count'].sum())
).reset_index(drop=True)
proportion_data
```

```python
def summarize_data(data, label):
    summary = data.groupby(['community', label]).size().reset_index(name='count')
    summary = summary[summary[label] == 'ADHD']
    summary[label] = summary[label].apply(lambda x: f"{x}{label[-1]}")
    return summary
```

```python
summary_data_ctrl_adhd4 = ctrl_subjects.groupby('community', observed=False).size().reset_index(name='count')
summary_data_ADHD1 = summarize_data(data, 'ADHD1')
summary_data_ADHD2 = summarize_data(data, 'ADHD2')
summary_data_ADHD3 = summarize_data(data, 'ADHD3')
summary_data_ADHD4 = summarize_data(data, 'ADHD4')
```

```python
summary_data_combined = pd.concat([summary_data_ctrl_adhd4, summary_data_ADHD1, summary_data_ADHD2, summary_data_ADHD3, summary_data_ADHD4], ignore_index=True)
```
