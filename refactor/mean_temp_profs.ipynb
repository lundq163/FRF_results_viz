```python
def remove_small_communities(df, community_column, threshold=100):
    community_counts = df[community_column].value_counts()
    mask = df[community_column].isin(community_counts[community_counts >= threshold].index)
    filtered_df = df[mask]
    return filtered_df

cd1 = remove_small_communities(cd1, 'community')
value_counts_cd1 = cd1['community'].value_counts()
print(value_counts_cd1)
```

```python
cd2 = remove_small_communities(cd2, 'community')
value_counts_cd2 = cd2['community'].value_counts()
print(value_counts_cd2)
```

```python
melted_cd1 = pd.melt(cd1, id_vars='community', var_name='metric', value_name='mean')
melted_cd2 = pd.melt(cd2, id_vars='community', var_name='metric', value_name='mean')
```

```python
melted_cd1
```

```python
melted_cd2
```

```python
grouped_cd1 = melted_cd1.groupby(['community', 'metric'])['mean'].mean().reset_index()
grouped_cd2 = melted_cd2.groupby(['community', 'metric'])['mean'].mean().reset_index()
```

```python
grouped_cd1
```

```python
grouped_cd2
```

```python
def calculate_mae(grouped_cd1, grouped_cd2, community1, community2):
    df1 = grouped_cd1[grouped_cd1['community'] == community1]
    df2 = grouped_cd2[grouped_cd2['community'] == community2]
    merged_df = pd.merge(df1, df2, on='metric', suffixes=('_cd1', '_cd2'))
    merged_df['abs_diff'] = (merged_df['mean_cd1'] - merged_df['mean_cd2']).abs()
    mae = merged_df['abs_diff'].mean()
    return mae
```

```python
community1 = 4
community2 = 5

mae = calculate_mae(grouped_cd1, grouped_cd2, community1, community2)
print(f"The MAE between community {community1} in grouped_cd1 (ARMS1) and community {community2} in grouped_cd2 (ARMS2) is {mae:.2f}")
```
